```
# set first row as column index and drop 0th row 
#
laliga.columns=laliga.iloc[0]
```

```
#drop team column
laliga.drop(['Team'],inplace=True,axis=1)
```

```
laliga.head(10)
```

```
# 1st way : replace dash with zero
dash_cols = ['Points','GamesPlayed','GamesWon','GamesDrawn','GamesLost','GoalsFor','GoalsAgainst','Champion',
            'Runner-up','Third','Fourth','Fifth','Sixth','T','BestPosition']
for col in dash_cols :
  '''
  This function
  pd.to_numeric(laliga[col],errors='coerce') will throw exception, here one more paramerter name 'error=coerce' 
  which will handle error and fill non-numeric element with NAN, once we get NAN then fillna will replace NAN value with 0
    '''
  laliga[col]=pd.to_numeric(laliga[col],errors='coerce').fillna(0).astype(int)
```

```
# 2nd way : to replace dash with zero 
# this function cannot be used where we dont know element which are going to replace 
# here this method doest not replace all dash that might be with some other unicode in csv.
#better we use ist way to fill non numeric number
laliga.replace('-',0)
laliga.replace('_',0)
```

```
#display those Team which having debut between 1930-1980
laliga[(laliga['Debut'].astype(str).str.split('-').str[0].astype(int)>1930) & (laliga['Debut'].astype(str).str.split('-').str[0].astype(int)<1980)]
```

```
#print top 5 having highest points
laliga.sort_values('Points',ascending=False).iloc[:5,:]
```

```
# writing function Goal_diff_count

def Goal_diff_count(goal_for,goal_against) :
  return goal_for-goal_against

```

```
col_goalFor=laliga.GoalsFor 
col_goalAgainst = laliga.GoalsAgainst
col_goalDiff=[] #intialize list for goal Diff 
length = col_goalFor.size
for i in np.arange(0,length,1) :
  col_goalDiff.append(col_goalFor.iloc[i]-col_goalAgainst.iloc[i]) # append result of substration into goalDiff list 
laliga['GoalsDiff']=col_goalDiff #append col_goalDiff into dataframe
#col_goalFor-col_goalAgainst
```

```
#find out team having maximum and minimum value based on newly appened GoalsDiff column 
maxTeam = laliga.loc[laliga.GoalsDiff.idxmax()]
maxTeam
```

```
minTeam = laliga.loc[laliga.GoalsDiff.idxmin()]
minTeam
```
