### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
#initialize list for different cat
#max list
mx=[];
#min list
mi=[];
#column name list
cols=[];
#range list
rg=[];
#only i want to extract numeric values
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
newdf = iris.select_dtypes(include=numerics)
#loop values get appended for each iteration at column level
for column in newdf :
  cols.append(column)
  cmin=newdf[column].min();
  cmax=newdf[column].max();
  mx.append(cmax)
  mi.append(cmin)
  rg.append(cmax-cmin)


data=[mx,mi,rg]
#then creating new dataframe
newRg=pd.DataFrame(data,index='max min range'.split(),columns=cols)
newRg
```
